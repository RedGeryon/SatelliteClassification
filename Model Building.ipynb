```python
import helper
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.svm import SVC
import random
import pandas as pd
import geopandas as gpd
import pickle
import os
import glob
```

```python
# Load our target and training examples
def load_data(ratio_threshold=.2, exclude_list=[]):
    '''We are going to load up our data and features. Each block group will have
    n sub_samples with corresponding binary feature. Var ratio_threshold will
    determine which class the test data will be in.
    
    Args:
        ratio_threshold (float): Below this threshold, target = 0, and above, feature = 1.
            inspecting the data, 0.2 will split the classes in half, but that artificial
            boundary may not corespond with a split that would visually identify a
            neighborhood as proportionately lower income.
            
    Returns:
        dataframe: Returns a features and target array to be split and trained.
    '''
    data_dir = './data/samples/'
    target_fp = './data/blockgroup_income_merged.shp'
    target_data = helper.load_shapefile(target_fp)
    
    if exclude_list:
        target_data = target_data[~target_data['GEOID'].isin(exclude_list)]
    
#     target_data = target_data[target_data['median_inc'].notnull()]
    
    data = []
    target = []
    region_class = [0 if x <= ratio_threshold else 1 for x in target_data['25k_ratio'].values ]
#     region_class = target_data['median_inc'].apply(int)
    block_GEOID = target_data['GEOID'].values
    for block, rclass in zip(block_GEOID, region_class):
        temp_locations = []
        curr_dir = data_dir + f'{block}/'
        search_criteria = '06*.pkl'
        fp = curr_dir + search_criteria
        temp_locations += glob.glob(fp)
        
        for file in temp_locations:
            data.append(pickle.load(open(file, "rb")))
            target.append(rclass)
    
    return np.array(data), np.array(target)
```

```python
def quick_model2(clf, thres=.20, exclude_list=[]):
    if exclude_list:
        images, target = load_data(ratio_threshold=thres, exclude_list=exclude_list)
    else:
        images, target = load_data(ratio_threshold=thres)
        
    X_train, X_test, y_train, y_test = train_test_split(images, target,
                                                        stratify=target,
                                                        test_size=0.20)
#     mean = np.mean(X_train)
#     sd = np.std(X_train)
#     X_train = np.divide(np.subtract(X_train, mean), sd)
#     X_test = np.divide(np.subtract(X_test, mean), sd)

    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    
    probs = clf.predict_proba(X_test)
    probs = probs[:, 1]
    auc = roc_auc_score(y_test, probs)
    score = clf.score(X_test, y_test)
    print('AUC: %.3f' % auc)
    # calculate roc curve
    fpr, tpr, thresholds = roc_curve(y_test, probs)
    # plot no skill
    plt.plot([0, 1], [0, 1], linestyle='--')
    # plot the roc curve for the model
    plt.plot(fpr, tpr, marker='.')
    # show the plot
    plt.show()
    
    f1_s = precision_recall_fscore_support(y_test, preds, average='macro')
    print(f'Precision, recall, f-score: {f1_s}')
    print(f'Classifier score is: {score}')
    return f1_s
    
   
exclude_list = ['060750231031', '060750328021', '060750332042', '060759802001', '060750126022', '060750129023', '060750158011',
              '060750126021', '060759805011', '060750354001', '060750251003', '060750352012', '060750105001', '060750105002',
               '060750607003', '060750127001', '060750126021', '060750615003']


tree = RandomForestClassifier(n_estimators=200, n_jobs=-1, min_samples_split=10)
quick_model2(tree, thres=0.174, exclude_list=exclude_list)
```

```python
# Use Randomized Search CV to tune hyperparams

from sklearn.model_selection import RandomizedSearchCV

def evaluate(clf, X_test, y_test):
    predictions = (clf.predict(X_test))
    probs = clf.predict_proba(X_test)
    probs = probs[:, 1]
    auc = roc_auc_score(X_test, probs)
    score = clf.score(X_test, y_test)
    print('AUC: %.3f' % auc)
    # calculate roc curve
    fpr, tpr, thresholds = roc_curve(y_test, probs)
    # plot random guess
    plt.plot([0, 1], [0, 1], linestyle='--')
    # plot the roc curve for the model
    plt.plot(fpr, tpr, marker='.')
    f1_s = precision_recall_fscore_support(y_test, preds, average='macro')
    print(f'Precision, recall, f-score: {f1_s}')
    print(f'Classifier score is: {score}')
    
    # Return precision, recall, f1 score, AUC, r2
    return f1_s[0], f1_s[1], f1_s[2], AUC, score
                   
param_grid = {
    'bootstrap': [True],
    'n_estimators': [int(x) for x in np.linspace(100, 500, 9)],
    'max_features': ['auto', 'sqrt'],
    'max_depth': [int(x) for x in np.linspace(10,310, 7)],
    'min_samples_split': [2,5,10,20],
    'min_samples_leaf': [1,2,4]
    }
                   
forest = RandomForestClassifier()
randomcv = RandomizedSearchCV(estimator=forest,
                              param_distributions=param_grid,
                              n_iter=20,
                              cv=2,
                              verbose=2,
                              n_jobs=-1)

images, target = load_data(ratio_threshold=.31, exclude_list=exclude_list)
randomcv.fit(images, target)
best = random.best_estimator_
random_precision, random_recall, random_f1, random_AUC, random_r2 = evaluate(forest, images, target)
```
